## Hiyerarşik Kümeleme

Amaç:
 1. sentetik veri seti oluştur ve hiyerarşik kümeleme uygula

Adımlar:
 1. veri seti oluşturma
 2. ham veri görselleştirme
 3. hiyerarşik kümeleme modeli kullanma
 4. veri noktalarının ait olduğu kümeleri belirle
 5. sonuçları görselleştir
 6. dendogram çizerek birleşme yapısını göster


```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
```

```python
# veri setini oluşturma
X, _ = make_blobs(n_samples= 300, centers= 4, cluster_std= 0.7, random_state= 42)
```

```python
# Ham veriyi görselleştirme
plt.figure(figsize = (4,3))
plt.scatter(X[:, 0],X[:,1], s=50, alpha=0.7, edgecolors="k")
plt.title("Ham veri")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

```python
# 3. hiyerarşik kümeleme modeli
model = AgglomerativeClustering(n_clusters= 4)
```

```python
# 4. veri noktalarının ait olduğu kümeleri belirle
labels = model.fit_predict(X)
```

```python
# 5. sonuçları küme renkleri ile görselleştir
plt.figure(figsize= (8,6))
plt.scatter(X[:,0], X[:,1], c=labels, cmap = "viridis", alpha=0.7, edgecolors="k")
plt.title("Hiyerarşik Kümeleme Sonucu")
plt.xlabel("X1")
plt.ylabel="X2"
plt.show()
```

```python
# 6. dendogram
baglanti_matrisi = linkage(X)

plt.figure(figsize = (20,5))
dendrogram(baglanti_matrisi)
plt.show()
```
