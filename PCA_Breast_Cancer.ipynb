##PCA WITH BREAST CANCER DATASET

Dataset:
 1. 569 satır, 30 özellik içeriyor
 2. hedef sınıflar malignant ve benign
 3. PCA ile bunu 30 özellikten 2 özelliğe indirgeyeceğiz


```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
breast_cancer = load_breast_cancer()
```

```python
x = breast_cancer.data
y= breast_cancer.target
```

```python
standard_scaler = StandardScaler()
x_std = standard_scaler.fit_transform(x)
```

```python
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_std)
x_pca.shape
```

```python
plt.figure()
for i in range(len(breast_cancer.target_names)):
  plt.scatter(x_pca[y == i, 0], x_pca[y == i, 1], label = breast_cancer.target_names[i])
plt.legend()






```
