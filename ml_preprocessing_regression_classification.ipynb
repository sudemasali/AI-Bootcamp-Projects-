```python
import pandas as pd
from sklearn.model_selection import train_test_split #machine learning python kütüphanesi sklearn'dür
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
#StandartScaler ölçeklendirme için gerekli olan sınıf(standardisation)
#MinMaxScaler Normalizasyon için gerekli olan sınıf
```

```python
df=pd.read_csv("musteri_verisi_ml_pratik.csv")
df.head()
```

```python
#missing value var mı bakıyoruz
df.isnull().sum()
```

```python
#kayıp verileri dropna ile siliyoruz
df_dropna=df.dropna()
df_dropna
```

```python
#sayısal sütunlardaki eksik verileri median ile doldurma
df_filled=df.copy()
sayisal_sutunlar=("yas","maas","deneyim_yili")

for sutun in sayisal_sutunlar:
  df_filled[sutun]=df_filled[sutun].fillna(df_filled[sutun].median())
```

```python
#kategorik sütundaki eksik veriyi mod ile doldur
df_filled["egitim"] = df_filled["egitim"].fillna(df_filled["egitim"].mode()[0])
df_filled
```

```python
#satın alıp almama bilgisini label encoding yapma
label_df = df_filled.copy()
label_encoder = LabelEncoder()
label_df["satin_aldi_bilgisi"] = label_encoder.fit_transform(label_df["satin_aldi"])
label_df
```

```python
#satın alıp almama bilgisini One Hot Encoding yapma
onehot_target_df = pd.get_dummies(df_filled["satin_aldi"], prefix="satin_aldi") #burda sonuç ayrı bir df olarak tutuluyor
onehot_target_df
```

```python
#One Hot Encoding ve Label Encoding için farklı bir örnek
onehot_df = pd.get_dummies(df_filled, columns = ["egitim"], drop_first=True)  #df içindeki sütunları yerinde dönüştürüyor
onehot_df["satin_aldi"] = label_encoder.fit_transform(onehot_df["satin_aldi"])
onehot_df
```

```python
#Standardizasyon ve Normalizasyona geçmek için elimizdeki veriyi x ve y şeklinde ikiye bölüyoruz
x = onehot_df.drop("satin_aldi" , axis=1) #features
y = onehot_df["satin_aldi"]  # target (satın alıp almama durumu) (tahmin etmek istediğimiz değer)
```

```python
#yalnızca sayısal sütunlarda standardizasyon ve normalizasyon yapacağız
sayisal_ozellikler=["yas","maas","deneyim_yili"]
```

```python
#standardizasyon(veriyi ortalaması 0 ve standart sapması 1 olacak şekilde dönüştürür)
X_standard = x.copy()
standard_scaler = StandardScaler()
X_standard[sayisal_ozellikler] = standard_scaler.fit_transform(X_standard[sayisal_ozellikler])
X_standard
```

```python
#Normalizasyon (değerleri 0 ile 1 arasına sıkıştırır)
X_normalized = x.copy()
minmax_scaler = MinMaxScaler()
X_normalized[sayisal_ozellikler] = minmax_scaler.fit_transform(X_normalized[sayisal_ozellikler])
X_normalized
```

```python
# Train test split ile veriyi eğitim ve test olarak ayıracağız
x_train, x_test, y_train, y_test = train_test_split(X_standard, y, test_size=0.2 , random_state=42)
print("x_train", x_train.shape)
print("x_test", x_test.shape)
print("y_train", y_train.shape)
print("y_test", y_test.shape)


```
