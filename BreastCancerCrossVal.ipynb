#breast cancer veri seti ile 10 fold cross validation implementasyonu

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold, StratifiedKFold, LeaveOneOut, cross_val_score
from sklearn.preprocessing import StandardScaler
```

```python
#veri setini yükle
X,y = load_breast_cancer(return_X_y= True)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

breast_cancer veri setindeki feature lar'ın ölçekleri birbirinden çok farklı olması sebebi ile veriyi standardize ettik. Aksi halde logistic regression iteratif çalışan bir model olduğu için bu durum öğrenmeyi zorlaştırırdı.

```python
model= LogisticRegression(max_iter=1000)
```

```python

kfold = KFold(n_splits = 10, shuffle = True, random_state= 42)
k_fold_scores = cross_val_score(model,X_scaled,y, cv = kfold)
k_fold_scores
```

```python
# stratified k-fold
stratified_kfold = StratifiedKFold(n_splits=5,shuffle=True,random_state=42)
stratified_kfold_scores = cross_val_score(model, X_scaled, y, cv=stratified_kfold,scoring="accuracy")
stratified_kfold_scores
```

```python

#leave one out
loo = LeaveOneOut()
loo_scores = cross_val_score(model, X_scaled, y, cv=loo, scoring="accuracy")
loo_scores
```
